```python
def replacer(df):
    cat = list(df.columns[df.dtypes=='object'])
    for i in df.columns:
        if i in cat:
            md = df[i].mode()[0]
            df[i] = df[i].fillna(md)
        else:
            mn = df[i].mean()
            df[i] = df[i].fillna(mn)
    print('Missing Values Replaced in dataframe')
```

```python
replacer(df)
```

```python
df.isna().sum()
```

Weight ~ Remaining Features

Step 5: Seperate X and Y(Weight)

```python
X = df.drop(columns=['id', 'Weight'])
Y = df[['Weight']]
```

```python
X.head()
```

```python
Y.head()
```

Preprocessing is always done on X features

```python
X_cat = X.select_dtypes(include='object')
X_con = X.select_dtypes(exclude='object')
```

```python
X_cat
```

```python
X_con
```

Categorical Features apply one hot encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(handle_unknown='ignore', 
                    sparse_output=False).set_output(transform='pandas')
X_cat_pre = ohe.fit_transform(X_cat)
X_cat_pre.head()
```

```python
X_cat_pre.to_csv('Ohe.csv', index=False)
```

Continouos features apply Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_con.head()
```

```python
scaler = StandardScaler().set_output(transform='pandas')
X_con_pre = scaler.fit_transform(X_con)
X_con_pre.head()
```

Combine Categorical and continuous features

```python
X_pre = X_con_pre.join(X_cat_pre)
X_pre.head()
```

```python
X_pre.to_csv('PreprocessedCars.csv', index=False)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.metrics import mean_absolute_percentage_error
def evaluate_model(model, x, y):
    ypred = model.predict(x)
    mse = mean_squared_error(y, ypred)
    rmse = mse**(1/2)
    mae = mean_absolute_error(y, ypred)
    mape = mean_absolute_percentage_error(y, ypred)
    r2 = r2_score(y, ypred)
    print(f'Mean Squared Error : {mse:.2f}')
    print(f'Root Mean Squared Error : {rmse:.2f}')
    print(f'Mean Absolute Error : {mae:.2f}')
    print(f'Mean Absolute Percentage Error : {mape:.4f}')
    print(f'R2 Score : {r2:.4f}')
```

```python
evaluate_model(model, xtrain, ytrain)
```

```python
evaluate_model(model, xtest, ytest)
```

R2 score is more than 80% for both train and test results hence it is a good model
